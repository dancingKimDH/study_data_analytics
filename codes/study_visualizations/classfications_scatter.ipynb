```python
class1_x_value = [0, 1, 2, 3, 4]
class1_y_value = [1, 3, 5, 7, 9]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=class1_x_value, y=class1_y_value)
```

```python
class2_x_value = [1.5, 2.5, 3.5, 4.5, 5.5]
class2_y_value = [3.5, 5.5, 7.5, 9.5, 11.5]
```

## 하나의 Plot에 표시

```python
#  두 클래스 변수를 한 plot에 표현
plt.scatter(x=class1_x_value, y=class1_y_value)
plt.scatter(x=class2_x_value, y=class2_y_value)
plt.show()
```

### Title, Label, and Legend (제목, 레이블, 범례)

```python
plt.scatter(x=class1_x_value, y=class1_y_value, label='Class One')
plt.scatter(x=class2_x_value, y=class2_y_value, label='Class Two')
plt.title('Classfications Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

plt.show()
```
