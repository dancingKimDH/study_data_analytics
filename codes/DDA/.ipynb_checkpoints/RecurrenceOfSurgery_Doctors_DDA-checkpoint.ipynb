### 목표변수: 수술시간
    - 설명변수:
                - 고혈압여부
                - 당뇨여부
                - 흡연여부
                - 체중
                - 골밀도

```python
import numpy as np
```

```python
# 수술시간 긴 순서대로 정렬
ds = df_ROS[['수술시간', '고혈압여부', '당뇨여부', '흡연여부', '체중', '골밀도']]
ds_list = ds.sort_values(by=['수술시간'], ascending=[False])
ds_list
```

```python
# hp = hypertension
ds_hp = ds_list[['수술시간', '고혈압여부']]
ds_hp.pivot_table(index='고혈압여부', values = '수술시간', aggfunc=np.mean)
# 따라서 고혈압이 있는 환자의 평균 수술시간이 더 길었음을 알 수 있다.
```

```python
# db = diabetes
ds_db = ds_list[['수술시간', '당뇨여부']]
ds_db.pivot_table(index='당뇨여부', values = '수술시간', aggfunc=np.mean)
# 당뇨 병력이 있는 환자의 경우 수술시간이 소폭 증가하였으나 통계적으로 유의미한 수치는 보이지 않았다.
```

```python
# sm = smoking
ds_sm = ds_list[['수술시간', '흡연여부']]
ds_sm.pivot_table(index='흡연여부', values = '수술시간', aggfunc=np.mean)
```
